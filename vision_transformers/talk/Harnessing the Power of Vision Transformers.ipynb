# How?
- Every word -> Embedding
- Represent in vector space
- The closer the "distance", the closer the words are
    - Eg: King, Throne ; Food, Dosa
<img src="images/embed.png" height="60%" width="60%">

- Weighted sum over all the embedding vectors 
- Weighted by their **normalized dot products**
- How related the words in the sequence are

```python
def return_vecs(list_of_words):
    return [model.get_vector(x) for x in list_of_words]
```

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from numpy import random
from numpy import dot
from scipy.special import softmax
import seaborn as sns
# define example
data = ["boy", "playing", "with", "ball"]
vecs = return_vecs(data)
words= vecs.copy()
sz = vecs[0].size
# generating the weight matrices
random.seed(42)
W_Q = random.randint(sz, size=(sz, sz))
W_K = random.randint(sz, size=(sz, sz))
W_V = random.randint(sz, size=(sz, sz))

# generating the queries, keys and values
Q = words @ W_Q
K = words @ W_K
V = words @ W_V

# scoring the query vectors against all key vectors
scores = Q @ K.transpose()

# computing the weights by a softmax operation
weights = softmax(scores / K.shape[1] ** 0.5, axis=1)

# computing the attention by a weighted sum of the value vectors
attention = weights @ V
```

```python
from transformers import AutoTokenizer, AutoModel, utils
from bertviz import model_view, head_view
from transformers import BertTokenizer, BertModel
utils.logging.set_verbosity_error()  # Suppress standard warnings

model_version = 'microsoft/xtremedistil-l12-h384-uncased'
model = BertModel.from_pretrained(model_version, output_attentions=True)
tokenizer = BertTokenizer.from_pretrained(model_version)
sentence_a = "Boy playing with ball"
sentence_b = "Girl playing with ball"
inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='pt')
input_ids = inputs['input_ids']
token_type_ids = inputs['token_type_ids']
attention = model(input_ids, token_type_ids=token_type_ids)[-1]
sentence_b_start = token_type_ids[0].tolist().index(1)
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list) 
```

```python

head_view(attention, tokens, sentence_b_start)
```
